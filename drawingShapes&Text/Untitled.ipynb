Drawing shapes

```python
import cv2
```

```python
import numpy as np
```

```python
blank = np.zeros((500,500,3), dtype='uint8')
```

```python
from matplotlib import pyplot as plt 
```

```python
plt.imshow(blank)
```

```python
blank[:] = 0,255,0
```

```python
blank
```

```python
plt.imshow(blank)
```

Drawing ractangle

```python
rect = cv2.rectangle(blank,(0,0),(250,250),(0,255,0), thickness=2)
```

```python
plt.imshow(rect)
```
